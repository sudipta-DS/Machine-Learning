```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("seaborn")
```

```python
x = np.arange(10)
y1 = x**2
y2 = (2*x) + 3

print(x)
print(y1)
print(y2)
```

## Line Chart

```python
plt.plot(x,y1,color='red',label='apple',marker='o')
plt.plot(x,y2,color='green',label='kiwi',linestyle='dashed',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of Fruits over time")
plt.legend()
plt.show()
```

```python
plt.style.available
```

## Scatter Plots

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y1,color='red',label='apple',marker='^')
plt.scatter(x,y2,color='green',label='kiwi',linestyle='dashed',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot-Price of Fruits over time")
plt.legend()
plt.show()
```

## Bar-Graphs

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,12,15],width=0.5,label='Current Year',tick_label=['Gold','Platinum','Silver'])
plt.bar(x_coordinates+0.25,[10,20,12],width=0.5,label='Previous Year')
plt.xlabel("Metals")
plt.ylabel("Price Over Time")
plt.ylim(0,40)
plt.legend()
plt.show()
```

## Pie Charts

```python
plt.style.use("seaborn")
subjects = ['chem','phy','Math','English']
weightage = [10,30,50,10]
plt.pie(weightage,labels=subjects,explode=(0,0,0,0.1),shadow=True,autopct='%1.1f%%')
plt.show()
```

## Histogram

```python
xsn = np.random.randn(100)
```

```python
print(xsn)
sigma = 5
u = 60
x= np.round((xsn*sigma)+u)
x1 = np.round((xsn*5)+40)
```

```python
print(x)
```

```python
plt.hist(x,alpha=0.8,label="Math")
plt.hist(x1,alpha=0.9,label="Physics")
plt.title("Histogram")
plt.xlabel("Number Range")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```
