```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
k = 5
color = ["green","red","blue","yellow","Orange"]

clusters = {}

for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    
    cluster = {
        "center" : center,
        "points" : points,
        "color"  : color[i]
    }
    clusters[i]=cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def AssignPointToClusters1(clusters):
    for ix in range(X.shape[0]):
        dist = []
        curr_x = X[ix]
        
        for kx in range(k):
            d = distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        current_cluster = np.argmin(dist)
        clusters[current_cluster]['points'].append(curr_x) 
def clusterUpdate1(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        if pts.shape[0] > 0:
            new_mean= pts.mean(axis=0)
            clusters[kx]['points']=[]
            clusters[kx]['center']=new_mean
        
def plotClusters1(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        uk = clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color='black',marker="*")
```

```python
AssignPointToClusters1(clusters)
plotClusters1(clusters)
clusterUpdate1(clusters)
```

```python
AssignPointToClusters1(clusters)
plotClusters1(clusters)
clusterUpdate1(clusters)
```
