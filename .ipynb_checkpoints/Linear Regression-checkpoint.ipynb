# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
print(X)
print(Y)
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.show()
```

```python
x = 0
lr = 0.1
error = []
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - (lr*grad)
    print(x)
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
plt.show()
```
