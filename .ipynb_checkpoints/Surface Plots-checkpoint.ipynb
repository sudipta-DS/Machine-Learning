```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap = 'rainbow')
plt.show()
```

```python
a1 = np.arange(-1,1,0.02)
b1 = a1

a1,b1 = np.meshgrid(a1,b1)
print(a1)
print(b1)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a1,b1,a1**2+b1**2,cmap = 'rainbow')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a1,b1,a1**2+b1**2,cmap = 'rainbow')
plt.show()
```
