```python
x = np.array([1,2,3,4])
y = x*2
print(y)
```

```python
pt.style.use('seaborn')
```

```python
pt.plot(x,x**2+x,label="Stock Price 1")
pt.plot(x,x**2,label="Stock Price 2")
pt.xlabel("Price")
pt.ylabel("The value")
pt.title("Stock Prediction Graph")
pt.legend()
pt.show()
```

```python
pt.scatter(x,x**2+x,label="Stock Price 1")
pt.scatter(x,x**2,label="Stock Price 2")
pt.xlabel("Price")
pt.ylabel("The value")
pt.title("Stock Prediction Scatter Plot")
pt.legend()
pt.show()
```
