```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
df = pd.DataFrame()
df['X'] = X.reshape(100)
df['y'] = y
df
```

```python
df['pred1'] = np.mean(df['y'])
df['res1'] = df['y']-df['pred1']
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1 = DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
dt1.fit(df['X'].values.reshape(100,1),df['res1'].values)
df['pred2'] = 0.265458 + 0.1*dt1.predict(df['X'].values.reshape(-1,1))
```

```python
dt2 = DecisionTreeRegressor(max_leaf_nodes=8)
df['res2'] = df['y']-df['pred2']
dt2.fit(df['X'].values.reshape(100,1),df['res2'].values)
df['pred3'] = df['pred2'] + 0.1*dt2.predict(df['X'].values.reshape(-1,1))
```
